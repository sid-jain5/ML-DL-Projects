```
X = cancer_df.drop(['target'], axis = 1)
```

```
X
```

```
y = cancer_df['target']
```

```
y
```

```
from sklearn.model_selection import train_test_split
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 5)
```

```
X_train
```

```
from sklearn.svm import SVC
```

```
from sklearn.metrics import classification_report, confusion_matrix
```

```
model = SVC()
```

```
model.fit(X_train, y_train)
```

```
predictions = model.predict(X_test)
```

```
predictions
```

```
cm = confusion_matrix(y_test, predictions)
```

```
sns.heatmap(cm, annot = True)
```

1.Wrong because data not scaled (Feature scaling)

2.C value can be changed which is the penalization on missclassifiaction (very high will lead to overfitting)

3.gamma value can be changed which is the distance from the boundary till which points are considered (high gamma means less dist which leads to overfittin)


```
min_train = X_train.min()
```

```
range_train = (X_train - min_train).max()
X_train_scaled = (X_train-min_train)/range_train
```

```
min_test = X_test.min()
range_test = (X_test - min_test).max()
X_test_scaled = (X_test-min_test)/range_test
```

```
model.fit(X_train_scaled, y_train)
```

```
new_predictions = model.predict(X_test_scaled)
```

```
cm2 = confusion_matrix(y_test, new_predictions)
```

```
sns.heatmap(cm2, annot=True)
```

```
print(classification_report(y_test, new_predictions))
```

```
param_grid = {'C' : [0.1, 1, 10, 100], 'gamma' : [1, 0.1, 0.01, 0.001], 'kernel' : ['rbf']}
```

```
from sklearn.model_selection import GridSearchCV
```

```
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 4)
```

```
grid.fit(X_train_scaled, y_train)
```

```
grid.best_params_
```

```
grid_predictions = grid.predict(X_test_scaled)
```

```
cm3 = confusion_matrix(y_test, grid_predictions)
```

```
sns.heatmap(cm3, annot = True)
```

```
print(classification_report(y_test, grid_predictions))
```
